```python
#Convert letters to ASCII values, A=65, B=66,...,Z=90

def str2num(pl): 
    pln=[]
    for x in pl:
        pln.append(ord(x)-65)
    return pln

def num2str(pln):
    return "".join([chr(int(x)+65) for x in pln])


def caesar(pl,shift=3):
    pln = str2num(pl)
    pln_shifted = [(x+shift)%26 for x in pln] #shift teh letters by 3 
    return num2str(pln_shifted )
```

### Baby step-Giant step 


We solve the discrete log problem 2^x ≡ 71 (mod 131) by the Baby Step-Giant Step method of Subsection 10.2.2 of the textbook. We take N = 12 since N^2 > p−1 = 130 and we form two lists. The first is 2^j (mod 131) for 0 ≤ j ≤ 11, and the second is 71 x 2^(-j) (mod 131) for 0 ≤ j ≤ 11:

```python
for i in range(0,12): 
    print (i, mod(2,131)^i) #The first list is 2^j (mod 131) for 0 ≤ j ≤ 11:
```

```python
for i in range(0,12):
    print (i, mod(71*mod(2,131)^(-12*i),131)) #the second list is 71 x 2^(-j) (mod 131) for 0 ≤ j ≤ 11:
```

```python
#The number 128 is on both lists, so we see that 2^7 ≡ 71 · 2^(−12*4) (mod 131). 
#Therefore 71=2^(7+4*12)=2^(55) (mod 131)

71==power_mod(2,55,131)
```

### El Gamal

```python
#We choose Alice's public key triple  
#Choose a large prime
p = next_prime(2^100); p
```

```python
#Find a primitive root mod p 
alpha=mod(primitive_root(p),p); alpha
```

```python
# We choose a random integer between 2 and p-2
# randint(a, b) returns a random integer in range [a, b], including both end points.

a=randint(2,p-2); a
```

```python
#Alice's pubic key is the triple (p, alpha, alpha^a)
(p,alpha,alpha^a)
```

```python
#Bob wants to send a message m=10^30 to Alice.
m=10^30; 
b=randint(2,p-2); # We choose a random integer between 2 and p-2
C=[alpha^b,m*alpha^(a*b)]; C #He computes C=[alpha^b, m*alpha^(a*b)] and sends this pair to Alice
```

```python
#Alice receives a ciphertex C=[alpha^b, m*alpha^(a*b)].
# To recover m she computes first (alpha^b)^(-a)=(alpha^b)^(p-1-a) (mod p) 
# and then (alpha^b)^(-a)m(alpha)^(a*b)=m (mod p)
mod((alpha^b)^(p-1-a)*m*alpha^(a*b),p)==m
```

### The elliptic curve El Gamal cryptosystem

- Public parameters for Alice and Bob: A large prime p, an elliptic curve E mod p, a point P ∈ E of high order n, and a point Q ∈ E such that Q=sP for some integer s. The smallest positive integer k such that kP=infinity is the order of P

- Secret key of Alice: The integer s such that s<n and Q=sP

- Encryption: For a message m ∈ E the ciphertext is (C1,C2) = (kP, kQ + m) for a randomly chosen k < n. (How Bob send a message to Alice)

- Decryption: m = C2 − sC1. (How Alice would decrypt the message sent by Bob)

```python
E107 = EllipticCurve(GF(107),[3,-2]) #y^2=x^3+3x-2 mod 107
```

```python
P = E107([23,6]); P.order()
```

```python
s=randint(1,107); s #Choose s ramdomly such that s< the order of P; This is the privake key of Alice
```

```python
Q = s*P;Q #The point Q
```

```python
#Encryption: Suppose we have the message m = (86, 25). Bob chooses a random number k 

k = randint(1,107);k
```

```python
m = E107([86,25]); C = [k*P,k*Q+m];C #ciphertext sent to Alice
```

```python
#To recover the plaintext, Alice simply computes:
C[1]-s*C[0]
```
